```python
# 모든 파일을 읽어서 하나의 데이터 프레임으로 만들기
df_list = []
for file in glob('D:/Workspace/02.DataAnalysis/07.Catogram/data/소상공인시장진흥공단_상가(상권)정보_20211231/*.csv'):
    gs = pd.read_csv(file, encoding='cp949')
    df_list.append(gs)

df = pd.concat(df_list)
df.shape
```
